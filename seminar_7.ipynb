```python
import matplotlib.pyplot as plt
x_points = range(0, 21)
y_points = [0.3, 0.6, 0.8, 0.9, 1.3, 0.5, 0.2, 0.3, 0.2, 0.3, 0.7, 1.8, 2.3, 1.7, 1.3, 2.6, 1.5, 1.4, 1.2, 0.5, 0.3]
plt.plot(x_points, y_points, "r")
plt.plot(x_points, y_points, "--")
plt.plot(x_points, y_points, "o")
# to comine all 3 do plt.plot(x_points, y_points, "r--o")
plt.title("A Fabulous Line Graph")
plt.xlabel("Abscissa")
plt.ylabel("Ordinate")
plt.grid()
plt.show()
```

```python
Student_marks = [48, 77, 62, 45, 92, 40, 84, 72]
min = 100
max = 0
for i in Student_marks:
  if i < min:
    min = i
for y in Student_marks:
  if y > max:
    max = y
print("Minimum:", min)
print("Maximum:", max)
Student_marks.reverse()
print("Reversed list:", Student_marks)
```
